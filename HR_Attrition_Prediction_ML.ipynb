```python
X=df.drop(columns="Attrition" , axis=1)
Y=df["Attrition"]
```

```python
print(X.shape)
print(Y.shape)
```

```python
X_train, X_test , Y_train , Y_test=train_test_split(X , Y , test_size=0.2 , stratify=Y , random_state=3)
```

```python
print(X.shape , X_train.shape , X_test.shape)
print(Y.shape , Y_train.shape , Y_test.shape )
```

```python
from pandas.core.arrays import sparse
encoder=OneHotEncoder(drop="first" , handle_unknown="ignore" , sparse_output=False)
```

```python
cat_cols = X.select_dtypes(include=["object"]).columns
num_cols = X.select_dtypes(exclude=["object"]).columns

```

```python
X_train_cat=encoder.fit_transform(X_train[cat_cols])
```

```python
X_test_cat=encoder.fit_transform(X_test[cat_cols])
```

```python
encoded_cols=encoder.get_feature_names_out(cat_cols)
X_train_cat=pd.DataFrame(X_train_cat, columns=encoded_cols)
X_test_cat=pd.DataFrame(X_test_cat , columns=encoded_cols)
```

```python
X_train_cat.index=X_train.index
X_test_cat.index=X_test.index
```

```python
scaler=StandardScaler()
```

```python
X_train_stand=scaler.fit_transform(X_train[num_cols])
```

```python
X_test_stand=scaler.transform(X_test[num_cols])
```

```python
scaled_cols=scaler.get_feature_names_out(num_cols)
X_train_stand=pd.DataFrame(X_train_stand , columns=scaled_cols)
X_test_stand=pd.DataFrame(X_test_stand , columns=scaled_cols)
```

```python
X_train_stand.head()
```

```python
X_train_stand.index=X_train.index
X_test_stand.index=X_test.index
```

```python
X_train_final=pd.concat([X_train_stand , X_train_cat] , axis=1)
X_test_final=pd.concat([X_test_stand , X_test_cat] , axis=1)
```

```python
X_train_final.head()
```

```python
print(X.shape , X_train_final.shape , X_test_final.shape)
```

```python
print(Y.shape , Y_train.shape , Y_test.shape)
```

```python
models=[LogisticRegression(max_iter=10000) , SVC() , RandomForestClassifier(random_state=1)]
```

```python
hyper_param={'log_reg':{
  'C':[1 , 5 , 10 , 20]
},

'svc_param':{
    'kernel':['linear' , 'rbf' , 'poly' , 'sigmoid'],
    'C':[1,5 , 10 , 20]
},

'random_param':{
    'n_estimators':[10 , 30 , 70 , 100]
}             }
```

```python
model_list=list(hyper_param.keys())
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_train_enc = le.fit_transform(Y_train)
y_test_enc  = le.transform(Y_test)
```

```python
hyper_param[model_list[0]]
```

```python
def model_evaluations(model_using , params):

  result=[]

  i=0

  for model in model_using:

    key=model_list[i]

    param=params[key]

    i+=1

    classifier=GridSearchCV(model , param , cv=5)

    classifier.fit(X_train_final , y_train_enc)

    result.append({'model_used':model ,
                   'best_param':classifier.best_params_
                   ,
                   'best_score':classifier.best_score_}
                  )
  result=pd.DataFrame(result , columns=['model_used' , 'best_param' , 'best_score'])

  return result
```

```python
model_evaluations(models, hyper_param)
```

```python
best_model_work=SVC(kernel='linear', C=10 , class_weight="balanced")
```

```python
best_model_work.fit(X_train_final , y_train_enc)
```

```python
X_train_pred=best_model_work.predict(X_train_final)
accuracy=accuracy_score(y_train_enc, X_train_pred)
precision=precision_score(y_train_enc, X_train_pred)
recall=recall_score(y_train_enc, X_train_pred)
f1=f1_score(y_train_enc, X_train_pred)
print("training accuracy =" , round((accuracy*100), 2) ,'%')
print("training precision =" , round((precision*100), 2) ,'%')
print("training recall =" , round((recall*100), 2) ,'%')
print("training f1 =" , round((f1*100), 2) ,'%')
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_train_enc, X_train_pred)
print(cm)
```
